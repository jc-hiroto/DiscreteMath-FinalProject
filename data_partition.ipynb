```python
partition = 128
p="parts\clicks_p"
for i in range(partition):
    path = p+str(i)+".csv"
    f = open(path, "w")
    f.write("session,timestamp,item_id,category\n")
    f.close()
```

```python
fo = open('yoochoose-data\yoochoose-clicks.dat', 'r') 
next(fo)
line = fo.readline()
count = 0
while line:
    sid = int(line.split(',', 1)[0])
    pid=sid%128
    print(sid,"=>",pid,end="\r")
    path = p+str(pid)+".csv"
    fw = open(path, "a")
    fw.write(line.strip()+"\n")
    fw.close()
    line = fo.readline()
    count += 1
fo.close()
```
