```python
import numpy as np
import pandas as pd
import datetime
```

```python
def dateparse(dt_str):
    return datetime.datetime.strptime(dt_str, '%Y-%m-%dT%H:%M:%S.%fZ')
```

```python
clicks = pd.read_csv("yoochoose-data\yoochoose-clicks.dat",parse_dates = [1], date_parser = dateparse, nrows=100000)
clicks.info()
# This dataset is extremely big, don't run it plz.
```

```python
buys = pd.read_csv("yoochoose-data\yoochoose-buys.dat",parse_dates = [1], date_parser = dateparse, nrows=100000)
buys.info()
```

```python
buy_items = buys.groupby('item_id')
buy_items.get_group(214537888).sum()
```

```python
click_items = clicks[['item_id']].groupby('item_id')
print(click_items.size())
```

```python
for id, group in click_items:
    print("ID: "+ str(id))
    click_count = group.count()[0]
    print("Clicked: "+str(click_count))
    print("\n")
```
