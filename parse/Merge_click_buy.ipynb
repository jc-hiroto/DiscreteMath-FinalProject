```python
import pandas as pd
import numpy as np
buys = pd.read_csv("buys_by_item.csv")
clicks = pd.read_csv("clicks_by_item.csv")
```

```python
res = pd.merge(clicks,buys, on='item_id', how='outer')
res['avg_price'] = np.where(res['quantity']==0,res['price'],res['price']/res['quantity'])
res['percentage'] = np.where(res['quantity']==0,0,res['quantity']/res['clicks']*100)
res=res.rename(columns={"price":"total_income"})
```

```python
res.to_csv('item_overview.csv',index=False)
```
