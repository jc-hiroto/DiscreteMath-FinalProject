```python
import numpy as np
import pandas as pd
import datetime

def dateparse(dt_str):
    return datetime.datetime.strptime(dt_str, '%Y-%m-%dT%H:%M:%S.%fZ')

buys = pd.read_csv("yoochoose-data/yoochoose-buys.dat",parse_dates = [1], date_parser = dateparse)
```

```python
buy_items = buys[['item_id','price','quantity']].sort_values('price').groupby('item_id')
buy_items.head(10)
```

```python
buy_items.sum().reset_index()[['item_id','price','quantity']].to_csv('buys_by_item.csv',index=False)
```
