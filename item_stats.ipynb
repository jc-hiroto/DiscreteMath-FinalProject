```python
import pandas as pd
import numpy as np

def overview():
    # Print the item overview data
    print("========== STATS OVERVIEW ==========")
    bTotal=res['quantity'].sum()
    cTotal=res['clicks'].sum()
    print('Total number of selled items: '+str(bTotal))
    print('Total clicks: '+str(cTotal))
    print('Percentage of buying: '+str((bTotal/cTotal)*100)+"%")
    print("\n")
    
def popular(disp = 5):
    # Get popular items data
    #disp = int(input("Display TOP ? items:"))
    print("========== TOP "+str(disp)+" MOST PURCHASED ==========")
    print(res.sort_values(by='clicks',ascending=False).head(disp).to_string(index=False))
    print("\n")
    print("========== TOP "+str(disp)+" MOST CLICKED ==========")
    print(res.sort_values(by='quantity',ascending=False).head(disp).to_string(index=False))
    print("\n")
    
def itemStat(item_id):
    sres=res.loc[res['item_id'] == item_id]
    if sres.empty:
        print("No Result! Change an item_id and try again.")
    else:
        print("========== ITEM STATS ==========")
        print(sres.to_string(index=False))
    print("\n")
```
