```python
import pandas as pd
import numpy as np
res = pd.read_csv("item_overview.csv")
res.head()
res.info()
```

```python
# Print the item overview data
bTotal=res['quantity'].sum()
cTotal=res['clicks'].sum()
print('Total number of selled items: '+str(bTotal))
print('Total clicks: '+str(cTotal))
print('Percentage of buying: '+str((bTotal/cTotal)*100)+"%")
```

```python
# Get popular items data
disp = int(input("Display TOP ? items:"))
print("========== Top "+str(disp)+" most bought items ==========")
print(res.sort_values(by='clicks',ascending=False).head(disp))
print("\n")
print("========== Top "+str(disp)+" most clicked items ==========")
print(res.sort_values(by='quantity',ascending=False).head(disp))
print("\n")
```

```python
sid=int(input("Enter item_id to search:"))
print("\n")
sres=res.loc[res['item_id'] == sid]
if sres.empty:
    print("No Result! Change an item_id and try again.")
else:
    print("========== SEARCH RESULT ==========")
    print(sres)
```
